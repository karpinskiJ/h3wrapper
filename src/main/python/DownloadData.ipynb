```python
def generate_path(id,dataset):
    if(id<10):
        return f"tl_rd22_0{id}_{dataset}.zip"
    else:
        return f"tl_rd22_{id}_{dataset}.zip"



def generate_input_roads(src_url,target_path,files_number,dataset):
    ids = list(range(10000,10000+files_number+1))
    urls = [f"{src_url}/tl_rd22_{id}_{dataset}.zip" for id in ids]
    paths = [f"{target_path}\\tl_rd22_{id}_{dataset}.zip" for id in ids ]
    return list(zip(urls,paths))

def download_and_unzip(input):
        urllib.request.urlretrieve(input[0],input[1])
        print(input[1])
        unzip(input[1])
    
```
