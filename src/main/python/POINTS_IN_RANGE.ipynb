```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .config("spark.executor.memory", "1g")\
        .config("spark.driver.memory", "4g")\
.appName('VIZ') \
 .master("local[*]")\
    .getOrCreate()
from utils import *
from sedona.spark import *
from pyspark.sql.functions import *
SedonaContext.create(spark)
```

```python
points = spark.read.parquet("data/POINTS_IN_RANGE").where(col("point_id")==1).withColumn("buffer",expr("ST_AsText(buffer)"))\
.select("lat","long","h3_index_shape","buffer").toPandas()
points_exact = spark.read.parquet("data/POINTS_IN_RANGE_ADD1").where(col("point_id")==1).withColumn("buffer",expr("ST_AsText(buffer)"))\
.select("lat","long","h3_index_shape","buffer").toPandas()
```

```python
m = folium.Map(location=[ 31.160, -88.452]) 
add_polygon_to_map(points,m,"buffer",color="blue",opacity=0.001, line_weight=3)
add_polygon_to_map(points,m,"h3_index_shape",color="red",opacity=0.1, line_weight=2)
add_circle_to_map(points,m,"lat","long",color="blue")
m
```
